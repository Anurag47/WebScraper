```python
data=review_soup.find('div',{'class':'swINJg _3nrCtb'}) #extracts the tag containing all the reviews
data
```

```python
url='https://www.flipkart.com'+url
url
```

```python
req_data=requests.get(url)
all_reviews=bs(req_data.content,'html.parser')
all_reviews
```

```python
pages=all_reviews.find_all('div',{'class':'_2zg3yZ _3KSYCY'}) #extracts all the pages url info
p=pages[0]   #selects the first page
p
```

```python
page=p.span #selects the <span> tag 
page
```

```python
tpage=page.get_text() #extracts the text
tpage
```

```python
total_pages=int(tpage.split()[-1]) #extracts the total pages of reviews we have
total_pages
```

```python
#this is the way to find the next page url 
next_url=all_reviews.find_all('a',{'class':'_3fVaIS'}) #extract the next url tag
next_url
```

```python
next_url[0].span.get_text() #we get to know if we have more pages to read
#Since it is the first page it will only have length 1(Next)
#Last page will have val = 'Previous' and length 1
#Other pages will have length 2 (Next and Previous)
```

```python
rating=[]
name=[]
heading=[]
desc=[]
print(url)

for i in range(0,total_pages): #we iterate through all the pages
    reviews=all_reviews.find_all('div',{'class':'ooJZfD _2oZ8XT col-9-12'})
    for rev in reviews:
        ratings=rev.find_all('div',{'class':'hGSR34 E_uFuv'}) #extracts all the ratings tags from the page
        ratings=[e.get_text() for e in ratings] #extracts the text part only
        rating.append(ratings) #appends the ratings of the page to list
        headings=rev.find_all('p',{'class':'_2xg6Ul'}) #extracts all the headings tags from the page
        headings=[e.get_text() for e in headings] #extracts the text part only
        heading.append(headings) #appends the heading of the page to list
        names=rev.find_all('p',{'class':'_3LYOAd _3sxSiS'}) #extracts all the name tags from the page
        names=[e.get_text() for e in names] #extracts the text part only
        name.append(names) #appends the names of the page to list
        des=rev.find_all('div',{'class':'qwjRop'}) #extracts all the description tags from the page
        des=[e.get_text() for e in des] #extracts the text part only
        des=[i[:-9:] for i in des] #removes the READ MORE from the text
        desc.append(des) #appends the desc of the page to list
    next_u=all_reviews.find_all('a',{'class':'_3fVaIS'}) #find the next page tag
    val=next_u[0].span.get_text()
    #we find the nxt page url
    if val=='Previous' and len(next_u)==1:
        print('completed')
        break
    elif val!='Next':
        next_url=next_u[1].get('href')
    elif val=='Next':
        next_url=next_u[0].get('href')
    #we parse the next page
    url_new='https://www.flipkart.com'+next_url
    req_data_review=requests.get(url_new)
    all_reviews=bs(req_data_review.content,'html.parser')
    print(url_new)
```
